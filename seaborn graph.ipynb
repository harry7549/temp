```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
print(sns.get_dataset_names()) #get a list of all datasets that are in-built in Seaborn
```

```python
df=pd.read_csv('demo.csv')
df
```

```python
plt.scatter(df.AGE,df.GENDER)
plt.show()
```

```python
#relplot
#The Seaborn Relational Plot (relplot) allows us to visualise how variables within a dataset relate to each other. 

sns.relplot(data=df, x="AGE", y="GENDER")
#relplot() function provides access to several different axes-level functions that show the relationship between two variables 
```

```python
sns.relplot(data=df, x="AGE", y="GENDER",  col="SALARY", col_wrap=5)
```

```python
#sns.relplot(data=df, x="AGE", y="GENDER",kind="line", ci=None)
sns.relplot(data=df, x="AGE", y="GENDER",kind="line")
```

```python
sns.pointplot(x="GENDER", y="AGE", data=df)
plt.show()
```

```python
sns.pointplot(x="AGE", y="GENDER", 
                   data=df, palette="Accent")
```

```python
#Joint Plot
#Joint Plot draws a plot of two variables with bivariate and univariate graphs. 



sns.jointplot(x='AGE', y='GENDER',data=df)
```

```python
#Joint Plot
#Joint Plot draws a plot of two variables with bivariate and univariate graphs. 



# Add regression line to scatter plot and kernel density estimate to histogram
sns.jointplot(x='AGE', y='SALARY', data=df, kind='reg')  #kind - hex,kde 

# y='SALARY' ama string value nahi ave tema float value use thai che 
```

```python
#Regplot
#Regplot is one of the functions in Seaborn that are used to visualize the linear relationship as determined through regression. 

ax = sns.regplot(x="AGE", y="SALARY", data=df)

# y='SALARY' ama string value nahi ave tema float value use thai che
```

```python
sns.set_style("whitegrid")

sns.boxplot(x="AGE", y="SALARY", data=df , width=1)

#data, x, y, hue, order, hue_order, orient, color, palette, saturation, width, dodge, fliersize, linewidth, whis, ax, **kwargs) 
# tame aya upar apela fun use kari sako
```

```python
#Heatmap
#A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colours. 


flights = df.pivot("SALARY", "AGE", "ID")
print(flights)
```

```python
sns.heatmap(flights,linewidths=.8,cmap="YlGnBu")
```

```python
corr=df.corr()
print(corr)
sns.heatmap(corr,annot=True,linewidths=.5,cmap="YlGnBu")


#  A correlation matrix is a table containing correlation coefficients between variables. Each cell in the table
#  represents the correlation between two variables. The value lies between -1 and 1.
```

```python
#Sequential Palette : one color only
# libraries

import numpy as np
 
# Create a dataset
df = pd.DataFrame(np.random.randn((10,10)), columns=["a","b","c","d","e","f","g","h","i","j"])

# plot using a color palette
sns.heatmap(df, cmap="YlGnBu")
plt.show()

sns.heatmap(df, cmap="Blues")
plt.show()

sns.heatmap(df, cmap="BuPu") #customize the colors in your heatmap with the cmap parameter of the heatmap() function in seaborn.
plt.show()

sns.heatmap(df, cmap="Greens")
plt.show()
```

```python
# create dataset
df = np.random.randn(20,20)

# plot heatmap
sns.heatmap(df, center=5)
plt.show()
```

```python
# Create a dataset
df = pd.DataFrame(np.random.random((10,10)), columns=["a","b","c","d","e","f","g","h","i","j"])

# color bar range between 0 and 0.5
sns.heatmap(df, vmin=0, vmax=0.5)
plt.show()

# color bar range between 0.5 and 0.7
sns.heatmap(df, vmin=0.5, vmax=0.7)
plt.show()
```
